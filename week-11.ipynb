```python
# score_stats.py
import numpy as np

def class_average(scores):
    return np.mean(scores)

def variance(scores):
    return np.var(scores)
```

```python
# analysis/analyzer.py
import pandas as pd

def subject_average(df):
    return df.mean()

def topper(df):
    return df.max()
```
